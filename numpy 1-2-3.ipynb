```python
# numpy 1 作业1，请对前文提到的Lotto程序进行三次运行，每次将结果存入列表，并确保第三次运行时所有结果以二维数组形式呈现为（3,6）的形状。
import numpy as np

all_results = [] # 存储三次运行结果的列表
# 运行三次程序
for i in range(3):
    np.random.seed(85 + i)  # 每次使用不同的种子，确保结果不同
    a = np.arange(1, 46)
    np.random.shuffle(a)
    lotto = a[:6]
    all_results.append(lotto)
    print(f'第{i+1}次运行结果: {lotto}')

# 将结果转换为(3,6)形状的二维数组
result_array = np.array(all_results)
print(f'\n所有结果组成的二维数组:')
print(result_array)
print(f'数组形状: {result_array.shape}')
```
